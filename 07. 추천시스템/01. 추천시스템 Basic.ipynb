## 추천 시스템 Basic

## 1. 추천 시스템이란?
---
추천시스템의 필요성 및 해결 하고자하는 문제는 무엇인지? 어떤 정보가 사용되는지 학습
### 1.1 개요
Serch vs Recommendation
- Search : 사용자가 의도를 가지고 물품을 찾는 query라는 키워드가 주어진다.(Pull 방식) 원하는 것을 추천하주는 정확도가 높은 방식
- Recommend : 사용자의 취미에 받게 추천을 해준다(push 방식)
- 추천시스템의 필요성
    - 과거에는 유저가 접할 수 있는 상품, 컨텐츠가 제한적이라 본인이 선택하는데 문제가 없었다.
    - 웹/모바일환경으로 다양한 상품, 컨텐츠가 등장하게 되며 정보가 풍요로워지며 인기 있는 것들만 인기있어짐
    - 정보를 찾는데 오랜 시간이 걸린다. 유저 또한 어떤 키워드로 찾아야 하는지 모를 수 있다.
    - Long-Tail 문제를 해결하는 것이 목표

- Log-Tail Recommendation 사례
    - 유튜브 동영상 추천
        - 조회수가 급상승한 영상이 등장한 경우, 해당 영상과 관련 되거나 해당 채널의 영상을 추천해준다(조회수가 적더라도 = Long tail)
    - SNS 친구 추천
        - 수천만 수십억 유저들 가운데 내가 알만한 사람들을 추천

### 1.2 사용 데이터
추천 시스템에서 사용하는 정보의 종류
1. 유저 관련 정보
- 유저 프로파일링
    - 추천 대상 유저에 관련된 정보를 구추갛며, 개발 유저 혹은 유저 그룹별로 추천
- 식별자(identifier)
    - 유저 ID, 디바이스 ID, 브라우저 쿠키
- 데모그래픽 정보
    - 성별, 연령, 지역, 관심사
- 유저 행동 정보
    - 페이지 방문 기록, 아이템 평가, 구매 등의 피드백 기록
2. 아이템 관련 정보
- 추천 아이템의 종류
    - 포탈 : 뉴스, 블로그, 웹툰 등 컨텐츠 추천
    - 광고/커머스 : 광고 소재, 상품 추천
    - 미디어 : 영화, 음악, 동영상 추천
- 아이템 프로파일링
    - 아이템 ID
    - 아이템의 고유 정보
        - 영화 : 장르, 배우, 감독, 영화 소개글, 개봉년도
        - 상품 : 카테고리, 브랜드, 출시일, 상품 이미지
        - 음악 : 아티스트, 작곡가, 장르, 음악 신호 그 자체
3. 유저 - 아이템 상호작용 정보
- 유저와 아이템의 상호작용 데이터
    - 유저가 오프라인 혹은 온라인에서 아이템과 상호작용 할 때 로그로 남음
    - 추천 시스템을 학습하는 데이터의 Feedback이 됨
- Explicit Feedback
    - 유저에게 아이템에 대한 만족도를 직접 물어보는 경우
    - ex) 유저의 영화에 대한 1~5점 사이의 평점
- Implicit Feedback
    - 유저가 아이템을 클릭하거나 구매한 경우
    - 이 경우의 데이터가 가장 많다
    - ex) 쿠팡에서 유저가 상품을 구매하면 implicit feedback = Y

### 1.3 문제 정의
- 추천시스템의 목적
    - 특정 유저에게 적합한 아이템을 추천한다 / 특정 아이템에 적합한 유저를 추천한다
    - 유저-아이템 상호 작용을 평가할 score 값이 필요함
    - 추천을 위한 score는 어떻게 구해지고 사용 될까
- 추천 문제 : 랭킹 또는 예측
    - 랭킹(Ranking) : 유저에게 적합한 아이템 Top K 개를 추천하는 문제
        - Top K 개를 선정하기 위한 기준 스코어가 필요하지만
        - 유저(X)가 아이템(Y)에 가지는 정확한 선호도를 구할 필요는 없다(내림차순만 하면됨)
        - 평가지표 : Precision@K, Recall@K, MAP@K , nDCG@K
    - 예측(Prediction) : 유저가 아이템을 가지 선호도를 정확하게 예측해야한다(평점 or 클릭/무개 확률)
        - Explicit Feedback : 유저가 내릴 평점 값을 예측
        - Implicit Feedback : 유저가 물품을 구매할 확률을 예측
        - 유저-아이템 행렬을 채우는 문제
        - 평가 지표 : MAE, RMSE, AUC

### 1.4 적용사례
- 넷플릭스 : 비슷한 영화를 추천
- 쿠팡 : 비슷한 상품들을 보여줌
- Google Ads : 광고 추천, 사용자에게 어떤 광고를 내보내줄지

## 2. 추천시스템의 평가 지표
---
### 2.1 개요
새로 적용한 추천 시스템 or 추천 모델의 성능 평가는 어떻게 해야 할까
- 비즈니스 / 서비스 관점
    - 추천 시스템 적용으로 인해 매출, PV의 증가
    - 추천 아이템으로 인한 유저의 CTR의 상승
- 품질 관점
    - 연관성(Relevance) : 추천된 아이템이 유저에게 관련이 있는가
    - 다양성(Diversity) : 추천된 Top-K 아이템에 얼마나 다양한 아이템이 추천되는가
    - 새로움(Novelty) : 얼마나 새로운 아이템이 추천되고 있는가
    - 참신함(Serendipity) : 유저가 기대하지 못한 뜻밖의 아이템이 추천되는가
    
### 2.2 Offline Test
Offline Test 란?
- 새로운 추천 모델을 검증하기 위해 가장 우선적으로 수행되는 단계
    - 유저로부터 수집한 데이터를 tarin/valid/test로 나누어 모델의 성능을 객관적인 지표로 평가
    - 보통 offline test에서 좋은 성능을 보여야 online 서빙에 투입되지만, 실제 서비스 상황에서는 다양한 양상을 보임 (serving bias 존재)
- 성능지표
    - 랭킹 문제 : Precesion@K, Recall@K, MAP@K, NDCG@K, HitRate
    - 예측 문제 : RMSE, MAE
    
#### Preicision/ReCall@K   
Precision@K
- 우리가 추천한 K개 아이템 가운데 실제 유저가 관심있는 아이템의 비율

Recall@K
- 유저가 관심있는 전체 아이템 가운데 우리가 추천한 아이템의 비율

예시)
- 추천한 아이템의 개수 = 5(k)
- 추천한 아이템 중 관심있는 아이템의 개수 : 2
- 유저가 관심있는 아이템 전체개수 : 3
- Precision@5 = 2/5
- Recall@5 = 2/3
        

#### Mean Average Precision(MAP) @ K
AP@K
- Precision@1 ~ PrecisionK 까지의 평균값
- Precision@K와 달리, 관련 아이템을 더 높은 순위에 추천할수록 점수가 상승함
$$AP@K = \frac{1}{m}\displaystyle\sum^{k}_{i=1}Precision@i$$

MAP@K
- 모든 유저에 대한 Average Precision 값의 평균
$$MAP@K = \frac{1}{|U|}\displaystyle\sum^{|U|}_{u=1}(AP@K)_u$$

#### Normalized Discounted Cumulative Gain(NDCG)
- 추천 시스템에 가장 많이 사용되는 지표 중 하나, 원래 검색(Information Retrieval)에서 등장한 지표
- Precision@K, MAP@K와 마찬가지로 Top K 리스트를 만들고 유저가 선호하는 아이템을 비교하여 값을 구한다
- MAP@K와 마찬가지로 추천의 순서에 가중치를 더 많이 두어 성능을 평가하므로 1에가 가까울수록 좋음
- MAP과 달리, 연관성을 이진(binary) 값이 아닌 수치로도 사용할 수 있기 떄문에 유저에게 얼마나 더 관련있는 아이템을 상위로 노출시키는지 알 수 있음

#### NDCG Formula
- CumulativeGain
    - 상위 K개 아이템에 대하여 관련도를 합한 것
    - 순서에 따라서 Discount 하지 않고 동일하게 더한 값
    $$CG_K=\displaystyle\sum^{K}_{i=1}rel_i$$

- Discounted Cumulative Gain
    - 순서에 따라 Cumulative Gain을 Discount 함
    $$DCG_K=\displaystyle\sum^{K}_{i=1}\frac{rel_i}{log_2(i+1)}$$
- IdealDCG
    - 이상적인 추천이 일어났을 때의 DCG값
    - 가능한 DCG 값 중에 제일 크다
    $$IDCG = \displaystyle\sum^K_{i=1}\frac{rel_i^{opt}}{log_2(i+1)}$$
- Normalized DCG
    - 추천 결과에 따라 구해진 DCG를 IDCG로 나눈 값
   $$NDCG = \frac{DCG}{IDCG}$$
   
### 2.3 Online Test
Online A/B Test란?  

- Offline Test에서 검증된 가설이나 모델을 이용해 실제 추천 결과를 서빙하는 단계
    - 추천 시스템 변경 전후의 성능을 비교하는 것이 아니라, 동시에 대조군(A)와 실험군(B)의 성능을 평가
    - 대조군과 실험군의 환경은 최대한 동일해야한다
    - 실제 서비스를 통해 얻어지는 결과를 통해 최종 의사결정이 이루어짐
    - 대부분 현업에서 사용하는 최종 지표는 모델 성능이 아닌 매출, CTR 등의 비즈니스/서비스 지표

## 3. 인기도 기반 추천
---
인기도 기반 추천이란?

### 3.1 개요
- 말그대로 가장 인기가 있는 아이템을 추천
    - 그렇다면 '인기 있다' 라는 정의는 무엇일까
- 인기도의 척도
    - 조회수, 평균, 평점, 리뷰 개소, 좋아요/싫어요 수
- 예시
    - 네이버 쇼핑 랭킹 순
    - 다음 뉴스, 댓글 추천
    - 레딧 Hot 추천
    
그럼 어떻게 스코어를 만들어야 할까?  

1. Most PoPular
    - 조회수가 가장 많은 아이템을 추천
    - 뉴스에 추천, 유저들은 다른 유저들이 많이 관심 가지는 핫한 이슈를 보고싶어함

2. Highly Rated
    - 평균 평점이 가장 높은 아이템을 추천
    - 평균 평점이 높은 맛집일수록 맛있는 음식을 제공하는 레스토랑으로 기대됨
    
### 3.2 Most Popular

#### Score Formula
- 가장 많이 조회된 뉴스를 추천 or 좋아요가 가장 많은 게시글을 추천
- 하지만 뉴스의 가장 중요한 속성은 최신성이다(시간도 중요)
    - 조금도 정교한 수식이 필요

#### Hacker News Formula
$$score = \frac{page views-1}{(age+2)^{gravity}}$$
- 시간이 지날수록 age가 점점 증가하므로 score는 작아짐
- 시간에 따라 줄어드는 score를 조정하기 위해 gravity라는 상수를 사용

#### Reddit Formula
$$ score = log_{10}(ups - downs)+\frac{sign(ups-downs)\cdot seconds}{45000}$$
- 첫번째 term은 popularity, 두번째 term은 포스팅이 게시된 절대 시간
    - 나중에 게시된 포스팅일수록 절대시간이 크기 때문에 더 높은 score를 가짐
    - log 함수를 줌으로써 초반 시간에 더 높은 가중치를 부여 해준다
- 첫번째 vote에 대해서 가장 높은 가치를 부여하며, vote가 늘어날 수록 score의 증가 폭이 작아짐
    - 오래된 포스팅일수록 상대적으로 아주 많은 vote가 있어야 높은 score를 가짐
    
### 3.3 Highly Rated

#### Score Formlua
- 가장 높은 평점을 받은 영화 혹은 맛집을 추천
- 신뢰할 수 있는 평점인가? 평가의 개수가 충분한가?
$$score = f(rating, \#\ of\ ratings)$$    

#### Steam Rating Formula
$$avg_rating = \frac{\# \  of \ postive \  reviews}{\# of\  reviews}$$

$$score = avg_ragint - (avg_rating - 0.5)\cdot 2^{-log(\# \ of \ reviews)}$$

- rating은 평균 값을 사용하되, 전체 review 개수에 따라 rating 보정
    - review 개수가 너무 많거나 적을 때는 0/5를 기준으로 보정을 해준다
- review의 개수가 아주 많을 경우 평균 rating과 거의 유사해짐

#### Movie Rating(steam 응용)
$$avg_rating = \frac{sum(rating)}{\# \ of\ reviews}$$
$$score = avg_rating - (avg_rating - 3.0)\cdot 2^{-log(\# \ of \ reviews)}$$
- 영화 평점은 positive, negative가 아닌 1.0 ~ 5.0의 rating을 사용함
    - 0.5 대신 1.0~5.0 사이 중앙값인 3.0을 사용
    - 3.0 대신 모든 평점 데이터의 평균값을 사용해도 된다
- 마찬가지로 review 개수가 많아질수록 score는 평균 rating에 가까워진다
